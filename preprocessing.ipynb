```python
import pandas as pd
```

```python
physicians_dtypes = {"zip": str, "grd_yr": str}
```

```python
pcp_df = physicians_original_df[physicians_original_df['pri_spec'].isin(primary_care_spec)]
pcp_df
```

```python
pcp_df.to_csv("mi_pcp_df.csv")
```

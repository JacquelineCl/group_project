```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Import data
file_path = "uszips.csv"
# Drop index
zip_data_df = pd.read_csv(file_path, index_col= 0)
zip_data_df.head(10)
```

```python
# Column types
zip_data_df.dtypes
```

```python
# Row count
zip_data_df.count()
```

```python
# Drop all states save Michigan
zip_data_df.drop(zip_data_df.index[zip_data_df['state_name'] != "Michigan"], inplace = True)
zip_data_df
```

```python
# Drop duplicates
zip_data_df.drop_duplicates()
```

```python
# Check for populations with 0
zip_data_df.sort_values(["population"], ascending=True)
```

```python
# Drop null population values (most are duplicates)
zip_data_df.drop(zip_data_df.index[zip_data_df['population'] == 0], inplace = True)
zip_data_df
```

```python
# Drop columns
zip_data_df.drop(columns=["lat", "lng", "state_id", "state_name", "parent_zcta", "density", "county_fips", "county_weights", 
                          "county_names_all", "county_fips_all", "imprecise", "military", "timezone"])
```

```python
# Output to CSV
zip_data_df.to_csv("zipcodes.csv")
```

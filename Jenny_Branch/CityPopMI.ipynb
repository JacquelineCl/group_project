```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Import data
file_path = "uszips.csv"
# Drop index
zip_data_df = pd.read_csv(file_path)
zip_data_df.head(10)
```

```python
# Column types
zip_data_df.dtypes
```

```python
# Row count
zip_data_df.count()
```

```python
# Drop duplicates
zip_data_df.drop_duplicates()
```

```python
# Check for populations with 0
zip_data_df.sort_values(["population"], ascending=True)
```

```python
# Drop population values equal to zero 
zip_data_df.drop(zip_data_df.index[zip_data_df['population'] == 0], inplace = True)
zip_data_df
```

```python
# Drop population if null (most are duplicates)
zip_data_df.drop(zip_data_df.index[zip_data_df['population'] == "NaN"], inplace = True)
zip_data_df
```

```python
# Rename zip column
zip_data_df.rename(columns={'zip': 'zipcode'})
```

```python
# Drop columns
zip_data_df.drop(columns=["lat", "lng", "state_id", "state_name", "parent_zcta", "density", "county_fips", "county_weights", 
                          "county_names_all", "county_fips_all", "imprecise", "military", "timezone"])
```

```python
# Output to CSV
zip_data_df.to_csv("zipcodes.csv")
```
